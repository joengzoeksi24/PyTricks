```python
# Required Libraries
from pyMultiobjective.algorithm import  optimized_multiobjective_particle_swarm_optimization
from pyMultiobjective.test_functions import dent_f1, dent_f2
from pyMultiobjective.util import graphs, indicators
```

```python
# Plot Solution - Scatter Plot
parameters = {
    'min_values': (-5, -5),
    'max_values': (5, 5),
    'step': (0.1, 0.1),
    'solution': [], 
    'show_pf': True,
	  'show_pts': True,
	  'show_sol': False,
	  'pf_min': True,  # True = Minimum Pareto Front; False = Maximum Pareto Front
	  'custom_pf': [], # Input a custom Pareto Front(numpy array where each column is an Objective Function)
    'view': 'notebook'
}
graphs.plot_mooa_function(list_of_functions = [dent_f1, dent_f2], **parameters)
```

```python
# Run
parameters = {
    'swarm_size': 25,
    'min_values': (-5, -5),
    'max_values': (5, 5),
	  'mutation_rate': 0.15,
    'iterations': 1500,
    'eta': 3,
    'verbose': True
}
sol = optimized_multiobjective_particle_swarm_optimization(list_of_functions = [dent_f1, dent_f2], **parameters)
```

```python
# Plot Solution - Scatter Plot
parameters = {
    'min_values': (-5, -5),
    'max_values': (5, 5),
    'step': (0.1, 0.1),
    'solution': sol, 
    'show_pf': True,
	  'show_pts': True,
	  'show_sol': True,
	  'pf_min': True,  # True = Minimum Pareto Front; False = Maximum Pareto Front
	  'custom_pf': [], # Input a custom Pareto Front(numpy array where each column is an Objective Function)
    'view': 'notebook'
}
graphs.plot_mooa_function(list_of_functions = [dent_f1, dent_f2], **parameters)
```

```python
# Plot Solution - Parallel Plot
parameters = {
    'min_values': (-5, -5), 
    'max_values': (5, 5), 
    'step': (0.1, 0.1), 
    'solution': sol, 
    'show_pf': True, 
	  'pf_min': True,   # True = Minimum Pareto Front; False = Maximum Pareto Front
	  'custom_pf': [],  # Input a custom Pareto Front(numpy array where each column is an Objective Function)
    'view': 'notebook'
}
graphs.parallel_plot(list_of_functions = [dent_f1, dent_f2], **parameters)
```

```python
# Plot Solution - Andrews Plot
parameters = {
    'min_values': (-5, -5), 
    'max_values': (5, 5), 
    'step': (0.1, 0.1), 
    'solution': sol, 
	  'normalize': True,
	  'size_x': 15,
	  'size_y': 15,
    'show_pf': True, 
	  'pf_min': True, # True = Minimum Pareto Front; False = Maximum Pareto Front
	  'custom_pf': [] # Input a custom Pareto Front(numpy array where each column is an Objective Function)
}
graphs.andrews_plot(list_of_functions = [dent_f1, dent_f2], **parameters)
```

```python
# Indicators
parameters = {
    'min_values': (-5, -5), 
    'max_values': (5, 5), 
    'step': (0.1, 0.1), 
    'solution': sol, 
	  'pf_min': True, # True = Minimum Pareto Front; False = Maximum Pareto Front
	  'custom_pf': [] # Input a custom Pareto Front(numpy array where each column is an Objective Function)
}
gd   = indicators.gd_indicator(list_of_functions = [dent_f1, dent_f2], **parameters)
gdp  = indicators.gd_plus_indicator(list_of_functions = [dent_f1, dent_f2], **parameters)
igd  = indicators.igd_indicator(list_of_functions = [dent_f1, dent_f2], **parameters)
igdp = indicators.igd_plus_indicator(list_of_functions = [dent_f1, dent_f2], **parameters)
ms   = indicators.ms_indicator(list_of_functions = [dent_f1, dent_f2], **parameters)
sp   = indicators.sp_indicator(list_of_functions = [dent_f1, dent_f2], **parameters)

parameters_h = {
    'solution': sol, 
	  'n_objs': 2,
	  'ref_point': [11, 11], # A Reference Point. If empty, an arbitrary Reference Point will be Used,
    'normalize': False
}
hypervolume = indicators.hv_indicator(**parameters_h)

# Print Results

print('GD   = ', gd)
print('GDP  = ', gdp)
print('IGD  = ', igd)
print('IGDP = ', igdp)
print('MS   = ', ms)
print('SP   = ', sp)
print('HYP  = ', hypervolume)
```
